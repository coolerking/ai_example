# ヒストグラム

縦軸に度数、横軸に階級をとった統計グラフの一種で、データの分布状況を視覚的に認識するために主に統計学や数学、画像処理等で用いられる。

## 画像のヒストグラム
縦軸に個数、横軸にRGBの各値[0, 256]を盗ったグラフ
グレースケールの場合は横軸は1つだけになる

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("src/Lena.jpg")
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
color_list = ["blue", "green", "red"]
for i,j in enumerate(color_list):
    # ヒストグラム bgrの順 マスク画像があるか＝None 
    hist = cv2.calcHist([img], [i], None, [256],[0,256])
    plt.plot(hist, color=j)
```

```python
# グレースケール
img_gray = cv2.imread("src/Lena.jpg", 0)
hist2 = cv2.calcHist([img_gray], [0], None, [256],[0,256])
plt.plot(hist2)
```
