# ヒストグラム均一化
ヒストグラムの累積度数(輝度値０から画素数を累積したもの)のグラフの傾きが一定になるように変換する処理です。こうすることによって、コントラストが悪かったり、明るさが偏っている画像の全体的なバランスを改善することが可能になります。

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# グレースケール
img = cv2.imread("src/Lena.jpg", 0)
hist = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist)
```

```python
img_eq = cv2.equalizeHist(img)
hist_e = cv2.calcHist([img_eq], [0], None, [256], [0,256])
plt.plot(hist_e)
```

```python
cv2.imshow("img", img)
cv2.imshow("eq", img_eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
